# **Background**

A company plans to conduct a survey to understand its market share across a country, region, and sub-region. To address this objective, a data analyst was hired to analyze and visualize the market share data. The results will support the company in making informed strategic decisions.

# **Problem Statement**

The company aims to conduct data analysis to better understand its market share without incurring high research costs. As data analysts, we will address the following key questions:

Which market segments perform best in each country, region, and sub-region?

Which products are the best-selling in each country, region, and sub-region?

Which products generate the highest sales volume across different geographical levels?

Which products produce the highest profit in each country, region, and sub-region?

**Discount Analysis**

```python
discount_sales = df.groupby('Discount')['Sales'].sum().reset_index()

print("Sales by Discount Level:")
for index, row in discount_sales.iterrows():
    discount = row['Discount']
    total_sales = row['Sales']
    print(f"Discount {discount * 100}%: {total_sales:.2f}")
```

**Impact of Discount on Sales**

```python
plt.figure(figsize=(15, 6))
plt.plot(discount_sales['Discount'], discount_sales['Sales'],
         marker='o', color='blue', linestyle='-', linewidth=2)

plt.xlabel('Discount')
plt.ylabel('Total Sales')
plt.title('Total Sales by Discount Level')
plt.grid(True)
plt.tight_layout()
plt.show()
```

Based on the chart, a 0% discount (no discount) still generates the highest overall sales.

Among discounted levels, the highest sales occur at a 20% discount, with total sales of 764,594.

Meanwhile, the lowest sales are recorded at a 45% discount, with total sales of only 5,484.

**Impact of Discount on Profit**

```python
discount_profit = df.groupby('Discount')['Profit'].sum().reset_index()

print("Profit by Discount Level:")
for index, row in discount_profit.iterrows():
    discount = row['Discount']
    total_profit = row['Profit']
    print(f"Discount {discount * 100}%: {total_profit:.2f}")
```

```python
plt.figure(figsize=(8, 6))
plt.plot(discount_profit['Discount'], discount_profit['Profit'],
         marker='o', color='blue', linestyle='-', linewidth=2)

plt.xlabel('Discount')
plt.ylabel('Profit')
plt.title('Profit by Discount Level')
plt.grid(True)
plt.tight_layout()
plt.show()
```

Based on the chart, the highest profit is generated with no discount (0%), reaching 320,987.

The lowest performance occurs at a 70% discount, resulting in a loss of -40,075.

Overall, discounts ranging from 30% to 80% lead to negative profit.

Meanwhile, the highest profit among discounted levels is at a 20% discount, generating 90,337 in profit.

**Correlation Analysis Between Variables**

```python
# Select variables to analyze
selected_variables = ['Sales', 'Quantity', 'Discount', 'Profit']

# Create a DataFrame with selected variables
selected_data = df[selected_variables]

# Calculate correlation matrix
correlation_matrix = selected_data.corr()

plt.figure(figsize=(10, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm',
            fmt='.2f', linewidths=0.5)

plt.title('Correlation Heatmap Between Variables')
plt.show()
```

The heatmap shows the correlation among Sales, Quantity, Discount, and Profit.

A strong positive correlation is observed between Sales and Profit, indicating that higher sales generally lead to higher profit.

Meanwhile, Discount and Profit show a weak or negative correlation, suggesting that higher discounts do not necessarily increase profitability.

```python
df['Order Date'] = pd.to_datetime(df['Order Date'])

yearly_sales = df.groupby(df['Order Date'].dt.year)['Sales'].sum().reset_index()

plt.figure(figsize=(15, 6))
plt.plot(yearly_sales['Order Date'], yearly_sales['Sales'],
         marker='o', color='blue', linestyle='-', linewidth=2)

plt.xlabel('Year')
plt.ylabel('Total Sales')
plt.title('Annual Sales Trend')
plt.grid(True)

max_sales = yearly_sales['Sales'].max()
min_sales = yearly_sales['Sales'].min()

plt.text(yearly_sales['Order Date'].iloc[yearly_sales['Sales'].idxmax()], max_sales,
         f'Max: {max_sales:.2f}', ha='left', va='bottom', fontsize=10)

plt.text(yearly_sales['Order Date'].iloc[yearly_sales['Sales'].idxmin()], min_sales,
         f'Min: {min_sales:.2f}', ha='left', va='top', fontsize=10)

plt.tight_layout()
plt.show()
```

The annual sales trend shows a decline from 2020 to 2021, with the lowest annual sales recorded at 470,532.

However, from 2021 to 2023, sales show a consistent increase, reaching the highest annual sales of 733,947.
