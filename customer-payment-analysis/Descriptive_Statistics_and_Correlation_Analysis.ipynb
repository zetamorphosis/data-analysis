# **Library**

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

# **Import and Load Data**

```python
orders = pd.read_csv('orders_dataset.csv')
orders
```

```python
order_items = pd.read_csv('order_items_dataset.csv')
order_items
```

```python
order_payments = pd.read_csv('order_payments_dataset.csv')
order_payments
```

# **Merge 3 Files and Clean the Data**

```python
# Merge main and supporting files based on order_id
merged_df = orders.merge(order_items, on='order_id', how='left')
merged_df = merged_df.merge(order_payments, on='order_id', how='left')
```

```python
# Remove rows with missing values in important columns
cleaned_df = merged_df.dropna(subset=['price', 'freight_value', 'payment_value'])
```

```python
# Remove duplicate rows
cleaned_df = cleaned_df.drop_duplicates()
```

```python
cleaned_df
```

```python
print("Data size after cleaning:", cleaned_df.shape)
```

# **Business Question**

**Is there a relationship between product price (price) and the payment amount (payment_value) made by customers?**

```python
# Scatter plot product price vs payment amount
plt.figure(figsize=(8,5))
sns.scatterplot(x=cleaned_df['price'], y=cleaned_df['payment_value'], alpha=0.3)

plt.title('Correlation Between Product Price and Payment Amount')
plt.xlabel('Product Price (price)')
plt.ylabel('Payment Amount (payment_value)')
plt.grid(True)
plt.show()
```

```python
# Calculate Pearson correlation
correlation1 = cleaned_df['price'].corr(cleaned_df['payment_value'])
print(f"Correlation between price and payment_value: {correlation1:.4f}")
```

**What is the distribution of payment methods (payment_type) most frequently used by customers?**

```python
# Calculate payment method frequency (percentage)
payment_counts = cleaned_df['payment_type'].value_counts(normalize=True) * 100

print("Percentage of payment method usage:")
print(payment_counts.round(2))
```

```python
plt.figure(figsize=(8,5))

ax = sns.countplot(data=cleaned_df,
                   x='payment_type',
                   order=payment_counts.index)

plt.title('Payment Method Distribution')
plt.ylabel('Number of Transactions')
plt.xlabel('Payment Type')
plt.grid(True)

for p in ax.patches:
    ax.annotate(f'{int(p.get_height())}',
                (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center',
                va='bottom')

plt.show()
```

**How strong is the impact of shipping cost (freight_value) on the total payment amount (payment_value)?**

**Scatter Plot: Shipping Cost vs Total Payment**

```python
plt.figure(figsize=(8,5))

sns.scatterplot(x=cleaned_df['freight_value'],
                y=cleaned_df['payment_value'],
                alpha=0.3)

plt.title('Correlation Between Shipping Cost and Payment Amount')
plt.xlabel('Shipping Cost (freight_value)')
plt.ylabel('Payment Amount (payment_value)')

plt.grid(True)
plt.show()
```

```python
# Calculate correlation
correlation2 = cleaned_df['freight_value'].corr(cleaned_df['payment_value'])
print(f"Correlation between freight_value and payment_value: {correlation2:.4f}")
```

**Business Question Conclusions**
1. Result: Pearson Correlation = 0.7395
There is a strong positive relationship between product price (price) and total payment amount (payment_value).
This means that higher product prices generally lead to higher total payments made by customers.

2. The majority of customers use credit cards as their primary payment method.

3. Result: Pearson Correlation = 0.3739
There is a moderate positive relationship between shipping cost (freight_value) and total payment amount (payment_value).
However, its influence is not as strong as the product price.

# **Correlation Between Numerical Variables**

```python
num_data = cleaned_df[['price', 'freight_value', 'payment_value']]
corr_matrix = num_data.corr()

print("Correlation Matrix:\n", corr_matrix)
```

```python
plt.figure(figsize=(8, 5))
sns.heatmap(corr_matrix, annot=True, cmap='YlGnBu')

plt.title('Correlation Heatmap of Numerical Variables')
plt.show()
```

# **Mean and Variance Analysis**

```python
mean_vals = num_data.mean()
var_vals = num_data.var()

print("Mean:\n", mean_vals)
print("\nVariance:\n", var_vals)

fig, axs = plt.subplots(1, 2, figsize=(12, 4))

sns.barplot(x=mean_vals.index, y=mean_vals.values, ax=axs[0])
axs[0].set_title('Mean')

sns.barplot(x=var_vals.index, y=var_vals.values, ax=axs[1])
axs[1].set_title('Variance')

plt.tight_layout()
plt.show()
```

**Average product price: 120.82**

On average, customers buy products worth about 121.

**Average shipping cost: 20.05**

Shipping is relatively small compared to product price.

**Average total payment: 172.69**

Customers typically pay around 173 per transaction.

his shows that most of the payment comes from the product price, not shipping.

**Price variance: 34,032.62**

**Shipping variance: 251.58**

**Payment variance: 71,605.63**

this means:

Payment value has the highest variance, meaning transaction amounts vary widely.

Product price also varies a lot, which explains why total payment varies greatly.

Shipping cost has very low variance, meaning shipping fees are relatively stable.
