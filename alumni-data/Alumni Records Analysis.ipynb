```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

import warnings
warnings.filterwarnings("ignore")
```

```python
# Load dataset
file_path = '/content/TC-ALUMNI2016.xlsx'
df = pd.read_excel(file_path, header=None)

# Deteksi heading dan subheading pada indeks 0-2
headings = df.iloc[0, :]  # Heading pada baris pertama (index 0)
sub_headings_1 = df.iloc[1, :]  # Subheading pada baris kedua (index 1)
sub_headings_2 = df.iloc[2, :]  # Subheading pada baris ketiga (index 2)

# Data mulai dari baris keempat (index 3)
data_content = df.iloc[3:, :]

# List untuk menyimpan data yang diproses
processed_data = []

# Variabel untuk menyimpan heading sebelumnya yang valid
previous_heading = None

# Iterasi untuk memproses heading, subheading, dan isi data
for i in range(len(headings)):
    heading = headings[i]
    subheading_1 = sub_headings_1[i]
    subheading_2 = sub_headings_2[i]

    # Jika heading tidak kosong, perbarui previous_heading
    if pd.notna(heading):
        previous_heading = heading

    # Ambil data isi dari kolom terkait
    column_data = data_content.iloc[:, i]

    # Gabungkan data ke dalam format yang mudah dianalisis
    for row_index, value in column_data.items():  # Ganti iteritems dengan items
        processed_data.append({
            "Heading": heading if pd.notna(heading) else None,
            "Subheading 1": subheading_1 if pd.notna(subheading_1) else None,
            "Subheading 2": subheading_2 if pd.notna(subheading_2) else None,
            "Parent Heading": previous_heading,
            "Data": value
        })

# Konversi data yang diproses menjadi DataFrame
df_processed = pd.DataFrame(processed_data)

# Simpan data ke file Excel
output_file_path = "processed_alumni_data_with_content.xlsx"
df_processed.to_excel(output_file_path, index=False)

print(f"Processed data with content saved to '{output_file_path}'")
```

```python
import pandas as pd

# List file path untuk setiap tahun
file_paths = [
    '/content/TC-ALUMNI2016.xlsx',
    '/content/TC-ALUMNI2017.xlsx',
    '/content/TC-ALUMNI2018.xlsx',
    '/content/TC-ALUMNI2019.xlsx',
    '/content/TC-ALUMNI2020.xlsx',
    '/content/TC-ALUMNI2021.xlsx',
    '/content/TC-ALUMNI2022.xlsx',
    '/content/TC-ALUMNI2023.xlsx'
]

# Fungsi untuk memproses data per file
def process_file(file_path):
    df = pd.read_excel(file_path, header=None)

    # Deteksi heading dan subheading pada indeks 0-2
    headings = df.iloc[0, :]  # Heading pada baris pertama (index 0)
    sub_headings_1 = df.iloc[1, :]  # Subheading pada baris kedua (index 1)
    sub_headings_2 = df.iloc[2, :]  # Subheading pada baris ketiga (index 2)

    # Data mulai dari baris keempat (index 3)
    data_content = df.iloc[3:, :]

    # List untuk menyimpan data yang diproses
    processed_data = []

    # Variabel untuk menyimpan heading sebelumnya yang valid
    previous_heading = None

    # Iterasi untuk memproses heading, subheading, dan isi data
    for i in range(len(headings)):
        heading = headings[i]
        subheading_1 = sub_headings_1[i]
        subheading_2 = sub_headings_2[i]

        # Jika heading tidak kosong, perbarui previous_heading
        if pd.notna(heading):
            previous_heading = heading

        # Ambil data isi dari kolom terkait
        column_data = data_content.iloc[:, i]

        # Gabungkan data ke dalam format yang mudah dianalisis
        for row_index, value in column_data.items():
            processed_data.append({
                "Heading": heading if pd.notna(heading) else None,
                "Subheading 1": subheading_1 if pd.notna(subheading_1) else None,
                "Subheading 2": subheading_2 if pd.notna(subheading_2) else None,
                "Parent Heading": previous_heading,
                "Data": value
            })

    return pd.DataFrame(processed_data)

# Proses setiap file dan simpan sebagai file Excel terpisah
for file_path in file_paths:
    year = file_path.split('-')[-1].split('.')[0]  # Ekstrak tahun dari nama file
    df_processed = process_file(file_path)
    output_file_path = f"processed_tc_alumni_{year}.xlsx"
    df_processed.to_excel(output_file_path, index=False)
    print(f"Processed data for year {year} saved to '{output_file_path}'")
```

# **Analisis Data**

```python
file_paths = [
    'processed_tc_alumni_ALUMNI2016.xlsx',
    'processed_tc_alumni_ALUMNI2017.xlsx',
    'processed_tc_alumni_ALUMNI2018.xlsx',
    'processed_tc_alumni_ALUMNI2019.xlsx',
    'processed_tc_alumni_ALUMNI2020.xlsx',
    'processed_tc_alumni_ALUMNI2021.xlsx',
    'processed_tc_alumni_ALUMNI2022.xlsx',
    'processed_tc_alumni_ALUMNI2023.xlsx'
]
```

**Menghitung jumlah mahasiswa berdasarkan Jurusan**

```python
for file_path in file_paths:
    year = file_path.split('ALUMNI')[-1].split('.')[0]
    df_alumni = pd.read_excel(file_path)
    jurusan_data = df_alumni[df_alumni['Parent Heading'] == 'Jurusan']
    jurusan_counts = jurusan_data['Data'].value_counts()

    print(f"\nJumlah mahasiswa berdasarkan Jurusan untuk tahun {year}:")
    for jurusan, count in jurusan_counts.items():
        print(f"- {jurusan}: {count} mahasiswa")
```

**Menghitung jumlah mahasiswa berdasarkan Fakultas per file**

```python
for file_path in file_paths:
    year = file_path.split('ALUMNI')[-1].split('.')[0]
    df_alumni = pd.read_excel(file_path)
    fakultas_data = df_alumni[df_alumni['Parent Heading'] == 'Fakultas']
    fakultas_counts = fakultas_data['Data'].value_counts()

    print(f"\nJumlah mahasiswa berdasarkan Fakultas untuk tahun {year}:")
    for fakultas, count in fakultas_counts.items():
        print(f"- {fakultas}: {count} mahasiswa")
```

**Jumlah Data Jika menurut anda pekerjaan anda saat ini tidak sesuai dengan pendidikan anda, mengapa anda mengambilnya? Jawaban bisa lebih dari satu**

```python
target_parent_heading = 'F16 - Jika menurut anda pekerjaan anda saat ini tidak sesuai dengan pendidikan anda, mengapa anda mengambilnya? Jawaban bisa lebih dari satu'

for file_path in file_paths:
    year = file_path.split('ALUMNI')[-1].split('.')[0]
    df_alumni = pd.read_excel(file_path)

    filtered_df = df_alumni[df_alumni['Parent Heading'] == target_parent_heading]
    filtered_yes = filtered_df[filtered_df['Data'] == 'Ya']

    comparison_counts = filtered_yes['Subheading 1'].value_counts()
    print(f"\nAlasan Mengambil Pekerjaan Tidak Sesuai Pendidikan - Tahun {year}")
    print("=" * 70)
    for alasan, count in comparison_counts.items():
        print(f"- {alasan}: {count} responden")
    print("=" * 70)
```

**Jenis Bidang Wirausaha**

```python
target_parent_heading = 'F11 - Apa jenis perusahaan/instansi/institusi tempat anda bekerja sekarang?'
target_subheading_1 = 'Berwirausaha/perusahaan sendiri'
target_subheading_2 = 'Wirausaha Di Bidang'

for file_path in file_paths:
    year = file_path.split('ALUMNI')[-1].split('.')[0]

    df_alumni = pd.read_excel(file_path)

    filtered_df = df_alumni[
        (df_alumni['Parent Heading'] == target_parent_heading) &
        (df_alumni['Subheading 1'] == target_subheading_1) &
        (df_alumni['Subheading 2'] == target_subheading_2)
    ]

    data_counts = filtered_df['Data'].value_counts()

    print(f"Data counts untuk tahun {year}:")
    print(data_counts)
    print("\n")
```
